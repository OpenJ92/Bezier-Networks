```python
def plot_tensor_T(T_):
    fig, axes = plt.subplots(2, 5, figsize = (16, 8))
    for i in range(axes.shape[0]):
        for j in range(axes.shape[1]):
            axes[i,j].matshow(T_[axes.shape[1]*i + j,:,:], cmap='tab20c')
```

```python
plot_tensor_T(W[0])
```
