The following is the equation that returns the dimensions of the tensor resulting from the 2D Convolution specified by the initial shape, dialation, padding, kernel size and stride. 

$$ f = X_{i, \textit{out}} = \frac{X_{i, \textit{in}} + 2\textit{ padding} -  \textit{ dialation}\big(\textit{ kernel} -1 \big) - 1}{\textit{ stride}} + 1 $$

Our main question is what is the result of repeated composition of this function $f$ and can we form the parameters in such a way that we land particularly on a desired size. ie.

$$ f(X_{i, \textit{in}}, \dots) = \frac{X_{i, \textit{in}}}{M} $$

and in a larger sense, can we pick a composition of $f_{\textit{parameters}}$ so that the resulting dimesions follow a particular linear or exponential decay. ie.

$$ f_n(f_{n-1}(\dots f_2( f_1( f_0 (X_{i, \textit{in}}, \dots) \dots) \dots) \dots ) \dots) = \frac{X_{i, \textit{in}}}{M} $$
$$ \text{where } Q(m) = f_m(f_{m-1}(\dots f_2( f_1( f_0 (X_{i, \textit{in}}, \dots) \dots) \dots) \dots) \dots) = \alpha m + X_{i, \textit{in}} \text{ or } X_{i, \textit{in}}\alpha^{m}$$

Lets see if we can use an inductive argument to determine the nth application of the Convolution opperator. Then we can programatically build out the architectures we want.

$$ X_{i, \textit{in}} - X_{i, \textit{out}} = C $$
$$ $$
$$ \frac{\textit{stride}X_{i, \textit{in}}}{\textit{stride}} - \frac{X_{i, \textit{in}} + 2\textit{ padding} -  \textit{ dialation}\big(\textit{ kernel} -1 \big) - 1 + \textit{stride}}{\textit{ stride}} = C$$
$$ $$
$$ \frac{\textit{stride}X_{i, \textit{in}} - X_{i, \textit{in}} - 2\textit{ padding} +  \textit{ dialation}\big(\textit{ kernel} - 1 \big) + 1 - \textit{stride}}{\textit{stride}} = C $$

```python
def apply_convolution(shape, dialation, padding, kernel_size, stride):
    x, y, x_d, y_d, x_p, y_p, x_k, y_k, x_s, y_s = *shape, *dialation, *padding, *kernel_size, *stride
    x_ = np.floor(((x + 2+x_p - x_d*(x_k - 1) - 1) / x_s) + 1)
    y_ = np.floor(((y + 2+y_p - x_d*(y_k - 1) - 1) / y_s) + 1)
    return x_, y_
```

```python
def func_(shape, dialation, padding, kernel_size, stride):
    del_x, del_y = np.array(shape) - np.array(apply_convolution(shape, dialation, padding, kernel_size, stride))
    return del_x, del_y
```

```python
def construct_parameter_tensor(layers, Q):
    R = np.stack([Q, Q], axis = 1)
    S = np.stack([R for layer in range(layers)], axis = 0)
    return S
```

```python
def construct_network(layers, init_shape, parameter_tensor):
    shapes = [init_shape]
    for layer in range(layers):
        new_shape = apply_convolution(shapes[layer], *[parameter_tensor[layer, i] for i in range(4)])
        shapes.append(new_shape)
    return shapes
```

```python
def construct_ConvNNSEQ(shape_init, shape_end, layers, function):
    # Paradigm:  Evaluate the function(monotonic decrease //for now) and build the kernel size from said function.
    #             Note the only parameter that's shanging is the kernel size
    
    if function == 'linear':
        tensor_shape = np.rint([shape_init + (shape_end - shape_init)*layer / layers for layer in range(layers + 1)]).astype('int')
    elif function == 'exponential':
        alpha = np.e ** (np.log(shape_end / shape_init) / layers)
        tensor_shape = np.rint([shape_init*(alpha**(layer)) for layer in range(layers + 1)]).astype('int')
    
    network = []
    for layer in range(layers):
        #print(tensor_shape[layer])
        network.append(nn.Conv2d(out_channels=tensor_shape[layer + 1][2], 
                                 in_channels=tensor_shape[layer][2], 
                                 kernel_size=tensor_shape[layer][:2] - tensor_shape[layer + 1][:2] + 1
                                )
                            )
        network.append(nn.LeakyReLU())
    return network, tensor_shape
```

```python
def construct_ConvTransposeNNSEQ(shape_init, shape_end, layers, function):
    # Paradigm:  Evaluate the function(monotonic decrease //for now) and build the kernel size from said function.
    #             Note the only parameter that's shanging is the kernel size
    
    if function == 'linear':
        tensor_shape = np.rint([shape_init + (shape_end - shape_init)*layer / layers for layer in range(layers + 1)]).astype('int')
    elif function == 'exponential':
        alpha = np.e ** (np.log(shape_end / shape_init) / layers)
        tensor_shape = np.rint([shape_init*(alpha**(layer)) for layer in range(layers + 1)]).astype('int')
    
    network = []
    for layer in range(layers):
        #print(tensor_shape[layer])
        network.append(nn.ConvTranspose2d(out_channels=tensor_shape[layer + 1][2],
                                          in_channels=tensor_shape[layer][2],
                                          kernel_size=tensor_shape[layer + 1][:2] - tensor_shape[layer][:2] + 1)
        )
        network.append(nn.LeakyReLU())
    return network, tensor_shape
    
```

```python
network, tensor_shape = construct_ConvNNSEQ(np.array([32, 32, 3]), np.array([4, 4, 30]), 8, 'linear')
network_, tensor_shape_ = construct_ConvTransposeNNSEQ(tensor_shape[-1], np.array([32, 32, 3]), 8, 'linear')

NN = nn.Sequential(*network)
NNTranspose = nn.Sequential(*network_)

data = torch.from_numpy(np.random.random_sample(size = (2, 3, 32, 32))).float()
```

```python
def construct_voxel_diagramEDIT2(tensor_shape):
    max_0, max_1 = np.max(tensor_shape[:,0]), np.max(tensor_shape[:,1])
    prisms = []
    for prism in tensor_shape:
        prism_ = np.ones(shape = (prism))
        diff_0, diff_1 = max_0 - prism[0], max_1 - prism[1]
        
        u_padding, l_padding = [np.zeros(shape = (prism[0], int(func(diff_0 / 2)), prism[2])) for func in [math.ceil, math.floor]]
        le_padding, r_padding = [np.zeros(shape = (int(func(diff_1 / 2)), prism[1] + diff_0, prism[2])) for func in [math.ceil, math.floor]]
        z_padding = np.zeros(shape = (max_0, max_1, 20))
        
        prism_ = np.concatenate([u_padding, prism_ ,l_padding], axis = 1)
        prism_ = np.concatenate([le_padding, prism_ ,r_padding], axis = 0)
        prism_ = np.concatenate([z_padding, prism_, z_padding], axis = 2)
        prisms.append(prism_)
        print(prism, max_0, max_1, prism_.shape)
    return np.concatenate(prisms, axis = 2) == 1
```

```python
voxels = construct_voxel_diagramEDIT2(np.concatenate([tensor_shape, tensor_shape_], axis = 0))
```
